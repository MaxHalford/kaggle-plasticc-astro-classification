```python
import pandas as pd

meta = pd.read_csv('data/training_set_metadata.csv')
meta['target'] = meta['target'].astype('category')
meta['is_galactic'] = meta['hostgal_photoz'] == 0

lc = pd.read_csv('data/training_set.csv')
lc.head()
```

```python
meta.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

def plot_features(features):

    agg = meta.join(features, on='object_id')

    for i, feature in enumerate(features.columns):
        
        sns.catplot(x='target', y=feature, col='is_galactic', data=agg, kind='box')
```

```python
import numpy as np

features = lc.groupby(['object_id', 'detected'])['mjd'].apply(np.ptp)

while isinstance(features.index, pd.MultiIndex):
    features = features.unstack()
    
features = features.add_prefix(f'{features.columns.name}_')

plot_features(features)
```
